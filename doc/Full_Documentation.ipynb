## Plot

<b>Argument Introduction</b>
<br>
<br>
<i>ordinary argument:</i>
<br>arg = return
<br>
<br>
<i>axes arguments:</i>
<br>arg = return
<br>arg = {'[i,j]':return}
<br>arg = {return:'[i,j]'}
<br>arg = {return:['[i,j]']}
<br>
<br>
<i>axes-plot arguments:</i>
<br>arg = return
<br>arg = {'plot':return}
<br>arg = {'[i,j]':{'plot':return}}
<br>arg = {'[i,j]':{return:'plot'}}
<br>arg = {'[i,j]':{return:['plot']}}
<br>arg = {'[i,j]':return}}
<br>arg = {return:'plot'}
<br>arg = {return:['plot']}
<br>
<br>
<i>axes-axislabel arguments:</i>
<br>arg = return
<br>arg = {'axislabel':return}
<br>arg = {'[i,j]':{'axislabel':return}}
<br>arg = {'[i,j]':{return:'axislabel'}}
<br>arg = {'[i,j]':{return:['axislabel']}}
<br>arg = {'[i,j]':return}}
<br>arg = {return:'axislabel'}
<br>arg = {return:['axislabel']}
<br>
<br>
<i>note:</i>
<br>- 1 row figure is using '[i]' axes argument, e.g. '[1]'
<br>- multi rows figure is using '[i,j]' axes argument, e.g. '[1,1]'
<br>- axes argument starts from 1 (*different from matplotlib which starts from 0)
